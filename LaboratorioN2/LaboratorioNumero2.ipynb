## Análisis de Complejidad: Algoritmos de Búsqueda No Informada

Para formalizar el análisis, definimos las siguientes variables teóricas del entorno:
* **$b$ (Factor de ramificación):** Máximo número de sucesores de un nodo (en nuestro laberinto $b \le 4$, ya que nos movemos Norte, Sur, Este, Oeste).
* **$d$ (Profundidad):** Profundidad del nodo objetivo más superficial.
* **$m$ (Profundidad máxima):** Longitud máxima de cualquier camino en el espacio de estados.
* **$C^*$**: Costo de la solución óptima.
* **$e$**: Costo mínimo de cada acción (en nuestro caso, $e = 1$).

### Tabla Comparativa Teórica

| Algoritmo | Complejidad Temporal | Complejidad Espacial | ¿Es Completo? | ¿Es Óptimo? |
| :--- | :--- | :--- | :--- | :--- |
| **Búsqueda en Anchura (BFS)** | $O(b^d)$ | $O(b^d)$ | Sí | Sí* |
| **Búsqueda en Profundidad (DFS)** | $O(b^m)$ | $O(b \cdot m)$ | Sí** | No |
| **Costo Uniforme (UCS)** | $O(b^{\lfloor C^*e \rfloor})$ | $O(b^{\lfloor C^*e \rfloor})$| Sí | Sí |

*\* BFS solo es óptimo si todos los costos de paso son idénticos (como en este laberinto).*

*\*\* DFS es completo solo en este caso ya que el espacio es de profundidad finita. De no serlo, DFS no se considera una solución completa.*

### Conclusiones del Laboratorio
1. **Rendimiento de Memoria (Espacial):** DFS es el más eficiente en memoria porque solo necesita almacenar un solo camino desde la raíz hasta la hoja actual ($O(b \cdot m)$). Por el contrario, BFS y UCS guardan la frontera completa de cada nivel, consumiendo exponencialmente más memoria ($O(b^d)$).
2. **Optimalidad:** DFS encontró rápidamente una solución (con solo 18 nodos expandidos frente a los 32 de BFS y UCS), pero no garantiza que sea el camino más corto. BFS sí garantiza la ruta más corta.
3. **Equivalencia BFS - UCS:** Al tener costos de paso constantes ($e= 1$), el término $(\lfloor C^*e \rfloor)$ se vuelve matemáticamente igual a $d$. Por lo tanto, en nuestro experimento, el costo de UCS fue exactamente igual al de BFS dónde ambos expandieron 32 nodos.

### Cual conviene usar en nuestro caso?
Debido a que nos encontramos en un contexto donde nuestro espacio es finito y bastante lineal; y nuestro objetivo es encontrar el objetivo rápido y consumiendo menos memoria, nuestra mejor opción es elegir DFS debido a que encuentra la solución expandiendo menos nodos. 

Cabe aclarar que hubo un **"factor suerte"** ya que el algoritmo hizo que el ultimo movimiento insertado en la pila fuera casualmente la dirección correcta hacia el objetivo.

En un contexto real BFS y UCS siempre va a encontrar el camino más corto aunque utilicen más memoria.
